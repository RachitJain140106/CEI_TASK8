Q1: Explain the roles of the Driver, Cluster Manager, and Executor in a Spark application.

A1). The Driver is the main part of a Spark application that creates jobs, schedules tasks, and collects the final results. The Cluster Manager manages the available resources and assigns them to the application. The Executors run the tasks on worker machines, process the data, and send the results back to the Driver.

Q2: How does Spark’s Lazy Evaluation strategy improve performance when chain-processing large datasets?

A2) Spark uses **Lazy Evaluation**, which means it does not execute transformations immediately. Instead, it records all the operations and runs them only when an action like `show()` or `count()` is called. This helps Spark optimize the execution plan using a DAG, reducing unnecessary work and improving performance on large datasets.


Q3: Write a Spark command to read a CSV file located at "data/source.csv", ensuring the first row is treated as a header and inferSchema is enabled.

```python
d = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("data/source.csv")

d.show()
```

Q4: What is the difference between CSV and Parquet in terms of storage (row-based vs. columnar) and why does it matter for performance?

A4) CSV is a row-based file format, while Parquet is a column-based format. CSV files are simple and easy to read but usually take more storage and are slower to process. Parquet files are compressed, store data more efficiently, and allow Spark to read only the required columns, making data processing much faster.

Q5: Given a DataFrame df, write a query to select the columns product_id and price where the category is 'Electronics'.

```python
df.filter(df.category == "Electronics") \
  .select("product_id", "price") \
  .show()
```

Q6: Write the code to "revise" a DataFrame by renaming the column old_name to new_name and casting the price column from a String to a Double.

```python
from pyspark.sql.functions import col

df = df.withColumnRenamed("old_name", "new_name") \
       .withColumn("price", col("price").cast("double"))

df.show()
```

Q7: How does Spark use the Lineage Graph (DAG) to provide fault tolerance if a worker node fails?

A7) Spark maintains a Lineage Graph (DAG) that records all the transformations applied to the data. If a worker node fails, Spark uses this lineage to recompute only the lost data instead of processing everything again. This makes Spark fault-tolerant and ensures reliable execution.

Q8: Write a query to filter a DataFrame df_orders for rows where the status is 'Completed' AND the amount is greater than 1000.

```python
df_orders.filter(
    (df_orders.status == "Completed") &
    (df_orders.amount > 1000)
).show()
```

Q9: Explain the concept of Predicate Pushdown in Parquet and how it affects the amount of data loaded into memory.

A9) Predicate Pushdown is an optimization technique used with Parquet files where Spark sends filter conditions to the storage layer. This allows only the required rows to be read into memory instead of scanning the entire dataset, which reduces I/O operations and improves performance.

Q11: What is the difference between Transformations and Actions? Provide two examples of each.

A11) Transformations are operations like filter() and select() that create a new DataFrame but are not executed immediately. Actions, such as show() and count(), trigger the execution of all pending transformations and produce the final result.

Examples:

Transformations: filter(), select()
Actions: show(), count()

Q12: Write the Spark command to load a Parquet file from "path/to/input", filter out any rows where user_id is null, and save the result as a CSV at "path/to/output".

```python
df = spark.read.parquet("path/to/input")

df.filter(df.user_id.isNotNull()) \
  .write \
  .mode("overwrite") \
  .option("header", "true") \
  .csv("path/to/output")
```

Q13: In Spark Architecture, what is the difference between Client Mode and Cluster Mode?

A13) In Client Mode, the Driver program runs on the user's local machine, making it suitable for development and testing. In Cluster Mode, the Driver runs inside the cluster, allowing the application to continue even if the user's machine disconnects, which makes it better for production environments.

Q14: Write a query to filter a dataset for rows where the region is 'North' OR the priority is 'High'.

```python
df.filter(
    (df.region == "North") |
    (df.priority == "High")
).show()
```

Q15: When exploring a dataset, why is it safer to use .show(5) instead of .collect() on a multi-terabyte dataset?

A15) Using .show(5) is safer because it displays only the first five rows without loading the entire dataset into the Driver's memory. On very large datasets, .collect() retrieves all the data to the Driver, which can consume a lot of memory and may even cause the application to fail.

##**Removing duplicates and normal forms**

##**Final code for clean dataset summarizing**

```python
import pandas as pd
df1 = pd.read_csv("Combined_dataset.csv")
# keep only useful columns (converting 1nf and 2nf)
df1 = df1[[
    "product_id", "title", "product_description", "rating", "ratings_count",
    "initial_price", "discount", "final_price", "currency",
    "seller_name", "category"
]]

# fill missing values with a string
df1 = df1.fillna("Unknown")

# remove duplicates
df1 = df1.drop_duplicates()

# formula for derived column as initial price is price and ratings_count signifies quantity
df1["total_amount"] = df1["initial_price"]*df1["ratings_count"]

# exporting to csv
df1.to_csv("clean_dataset.csv")
df1
```
